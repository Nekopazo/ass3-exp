## 3) k-性能曲线 + 标准差误差带（分 DR）

```python
for dr_name in sorted(summary_df['dr'].unique()):
    sub = summary_df[summary_df['dr'] == dr_name].copy()
    backbones = sorted(sub['backbone'].unique())
    fig, axes = plt.subplots(len(backbones), 1, figsize=(14, 4.5 * len(backbones)), sharex=True)
    if len(backbones) == 1:
        axes = [axes]

    for ax, b in zip(axes, backbones):
        sub_b = sub[sub['backbone'] == b]
        for clf_name in sorted(sub_b['clf'].unique()):
            s = sub_b[sub_b['clf'] == clf_name].sort_values('k_num')
            ax.plot(s['k_num'], s['macro_f1_mean'], marker='o', label=clf_name)
            ax.fill_between(
                s['k_num'],
                s['macro_f1_mean'] - s['macro_f1_std'],
                s['macro_f1_mean'] + s['macro_f1_std'],
                alpha=0.15
            )
        ax.set_title(f'{dr_name} | {b}')
        ax.set_ylabel('Macro-F1')
        ax.set_xticks(sorted(sub_b['k_num'].unique()))
        ax.legend(loc='best')

    axes[-1].set_xlabel('k')
    plt.tight_layout()
    show_and_save(f'curve_macro_std_{dr_name}')
```

## 4) Pareto 前沿图（Accuracy vs Macro-F1）

```python
pareto_df = summary_df.copy()

def pareto_front(df, x='accuracy_mean', y='macro_f1_mean'):
    pts = df[[x, y]].to_numpy()
    is_pareto = np.ones(len(pts), dtype=bool)
    for i, p in enumerate(pts):
        if not is_pareto[i]:
            continue
        dominates = (pts[:, 0] >= p[0]) & (pts[:, 1] >= p[1]) & ((pts[:, 0] > p[0]) | (pts[:, 1] > p[1]))
        dominates[i] = False
        if dominates.any():
            is_pareto[i] = False
    return is_pareto

pareto_df['is_pareto'] = pareto_front(pareto_df)

plt.figure(figsize=(10, 8))
sns.scatterplot(data=pareto_df, x='accuracy_mean', y='macro_f1_mean', hue='clf', style='dr', size='k_num', sizes=(40, 220), alpha=0.75)
front = pareto_df[pareto_df['is_pareto']].sort_values('accuracy_mean')
plt.plot(front['accuracy_mean'], front['macro_f1_mean'], color='black', linewidth=2, label='Pareto front')
plt.title('Pareto Front: Accuracy vs Macro-F1')
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left')
plt.tight_layout()
show_and_save('pareto_accuracy_macro')

display(front[['backbone','dr','k_num','clf','accuracy_mean','macro_f1_mean']].sort_values(['macro_f1_mean','accuracy_mean'], ascending=False))
```

## 5) 热力图矩阵（Backbone 固定，DR×k 与 Classifier）

```python
tmp = summary_df.copy()
tmp['dr_k'] = tmp['dr'] + '-' + tmp['k_num'].astype(str)

for b in sorted(tmp['backbone'].unique()):
    sub = tmp[tmp['backbone'] == b]
    piv = sub.pivot_table(index='dr_k', columns='clf', values='macro_f1_mean', aggfunc='mean')
    order = sorted(piv.index, key=lambda x: (x.split('-')[0], int(x.split('-')[1])))
    piv = piv.loc[order]

    plt.figure(figsize=(8, max(6, 0.45 * len(piv))))
    sns.heatmap(piv, annot=True, fmt='.3f', cmap='YlGnBu')
    plt.title(f'Macro-F1 Heatmap | {b}')
    plt.xlabel('Classifier')
    plt.ylabel('DR-k')
    plt.tight_layout()
    show_and_save(f'heatmap_macro_{b}')
```

## 6) 稳定性改为变异系数（CV）与 Top-3 可视化

定义：`CV = std / mean`。
主排序规则：先 `macro_f1_mean` 降序，再 `cv_macro` 升序。


```python
eps = 1e-12
cv_df = summary_df.copy()
cv_df['cv_macro'] = cv_df['macro_f1_std'] / (cv_df['macro_f1_mean'] + eps)
cv_df['cv_acc'] = cv_df['accuracy_std'] / (cv_df['accuracy_mean'] + eps)

# 主排序：性能优先，稳定性次优
cv_ranked = cv_df.sort_values(['macro_f1_mean', 'cv_macro'], ascending=[False, True]).reset_index(drop=True)
top3 = cv_ranked.head(3).copy()

display(top3[['backbone','dr','k_num','clf','macro_f1_mean','macro_f1_std','cv_macro','accuracy_mean','accuracy_std','cv_acc']])

# 导出论文用的Top-3稳定配置表
top3.to_csv(RESULTS_DIR / 'summary' / 'paper_top3_cv.csv', index=False)

# 用 fold 级数据只画 Top-3 的分布图
fold_tmp = fold_df.copy()
fold_tmp['config'] = fold_tmp['backbone'] + ' | ' + fold_tmp['dr'] + '-' + fold_tmp['k_num'].astype(str) + ' | ' + fold_tmp['clf']
top3['config'] = top3['backbone'] + ' | ' + top3['dr'] + '-' + top3['k_num'].astype(str) + ' | ' + top3['clf']
top3_cfgs = top3['config'].tolist()
plot_top3 = fold_tmp[fold_tmp['config'].isin(top3_cfgs)].copy()

# 保持与top3表一致的顺序
plot_top3['config'] = pd.Categorical(plot_top3['config'], categories=top3_cfgs, ordered=True)

fig, axes = plt.subplots(1, 2, figsize=(12, 4.5))
sns.boxplot(data=plot_top3, x='config', y='macro_f1', ax=axes[0])
axes[0].set_title('Top-3 Macro-F1 Distribution')
axes[0].set_xlabel('Config')
axes[0].set_ylabel('Macro-F1')
axes[0].tick_params(axis='x', rotation=25)

sns.boxplot(data=plot_top3, x='config', y='accuracy', ax=axes[1])
axes[1].set_title('Top-3 Accuracy Distribution')
axes[1].set_xlabel('Config')
axes[1].set_ylabel('Accuracy')
axes[1].tick_params(axis='x', rotation=25)

plt.tight_layout()
show_and_save('cv_top3_distribution')

# Top-3 CV柱状图（论文可直接放）
plot_cv = top3[['config','cv_macro','cv_acc']].copy()
plot_cv = plot_cv.melt(id_vars='config', var_name='metric', value_name='cv')

plt.figure(figsize=(8.5, 4.5))
sns.barplot(data=plot_cv, x='config', y='cv', hue='metric')
plt.title('Top-3 Coefficient of Variation (lower is better)')
plt.xlabel('Config')
plt.ylabel('CV')
plt.xticks(rotation=25, ha='right')
plt.tight_layout()
show_and_save('cv_top3_bar')
```

## 7) 分类器两两胜率（按同一 backbone+dr+k+fold 比较）

```python
base_cols = ['fold', 'backbone', 'dr', 'k_num']
pivot = fold_df.pivot_table(index=base_cols, columns='clf', values='macro_f1')
clfs = sorted(fold_df['clf'].unique().tolist())

win = pd.DataFrame(0.0, index=clfs, columns=clfs)
for a in clfs:
    for b in clfs:
        if a == b:
            win.loc[a, b] = 0.5
            continue
        valid = pivot[[a, b]].dropna()
        if len(valid) == 0:
            win.loc[a, b] = np.nan
        else:
            win.loc[a, b] = (valid[a] > valid[b]).mean()

plt.figure(figsize=(7, 6))
sns.heatmap(win, annot=True, fmt='.2f', cmap='coolwarm', vmin=0, vmax=1)
plt.title('Pairwise Win Rate (Macro-F1)')
plt.tight_layout()
show_and_save('pairwise_winrate_clf')

display(win)
```
